### **Mutual Information**
In probability theory, Mutual Information between 2 **random** variables is a **measure of mutual dependence** between the two variables. ### **KL Divergence**
Denoted as **D_KL (P || Q)**, this is a type of statistical distance => a measure of how much an approximating probability distribution *Q* is different for a true probability distribution *P*, and can be defined as:



Consider two probability distributions, a true P and an approximating Q. Often, P represents the data, the observations, or a measured probability distribution and distribution Q represents instead a theory, a model, a description, or another approximation of P. The Kullback–Leibler divergence is then interpreted as the average difference of the number of bits required for encoding samples of P using a code optimized for Q rather than one optimized for P.

### **Probability Distribution**
This is specifically defined as a function that gives the probabilities of occurrence of possible events. In other words, this is a randomized phenomenon in terms of the sample space given and the probabilities of said events occurring. 

In this, each sample space is going to have a probability distribution. For instance, if *X* is used to denote the outcome of a coing toss, then the probability distribution of X would be *0.5*. These probability distributions are used to compare the relative occurrence of many different random values.

### **Mutual**
Shared. Involving two (or more) entities equally. This means that there theoretically must be something in common between two variables, **most likely being the information they share**.

### **Dependency**
A relationship where one variable is influence by or related to another variable. If two varaibles are dependent, then **we can assume that knowing about one will give information about the other**.

### **Divergence**
A measure to determine how different two things are. Statistically, it refers to how one probability distribution differs from another.

### **Entropy**
A measure of uncertainty or randomness within a system or variable. the question it specifically answers is **"How unpredictable is this variable?"**.
